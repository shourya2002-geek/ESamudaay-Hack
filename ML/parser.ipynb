```python
import json
import requests
from collections import defaultdict
import re
import pandas as pd
```

```python
class MakeApiCall:

    def get_data(self, api):
        response = requests.get(f"{api}")
        if response.status_code == 200:
            print("sucessfully fetched the data")
            self.formatted_print(response.json())
        else:
            print(
                f"Hello person, there's a {response.status_code} error with your request")

    def get_user_data(self, api, parameters):
        response = requests.get(f"{api}")
        if response.status_code == 200:
            self.formatted_print(response.json())
        else:
            print()

    def formatted_print(self, obj):
        if len(obj)<1:
            pass
        else:
            for j in range(len(obj)):
                if obj[j]['sku_id']!=None:
                    self.sku.append(obj[j]['sku_id'])
                else:
                    self.sku.append("NA")
                    
                if obj[j]['product_name']!=None:
                    self.pdt.append(obj[j]['product_name'])
                else:
                    self.pdt.append("NA")
                    
                if obj[j]['failure_reasons']!=None:
                    self.failure.append(obj[j]['failure_reasons'])
                else:
                    self.failure.append("NA")
                    
                if obj[j]['business_name']!=None:
                    self.name.append(obj[j]['business_name'])
                else:
                    self.name.append("NA")
                    
    def __init__(self, api):
        # self.get_data(api)
        self.sku=[]
        self.name=[]
        self.failure=[]
        self.pdt=[]
        parameters = {
            "username": "kedark"
        }
        for i in api:
            self.get_user_data(i, parameters)
        print(self.sku)
        print(self.pdt)
        print(self.failure)
        print(self.name)
        self.df=pd.DataFrame(columns=[['sku_id'],['product_name'],['failure_reasons'],['business_name']])
        self.df['sku_id']=self.sku
        self.df['product_name']=self.pdt
        self.df['failure_reasons']=self.failure
        self.df['business_name']=self.name
        self.df.to_csv('data.csv') 
        print(self.df.head())
```
